```python
import json
import matplotlib.pyplot as plt

# Đọc dữ liệu từ file JSON
with open('/media/hoangtv/0f9d3910-0ff9-406c-92e1-c2c8170ca6f4/Dat/lip_av/mobilenetv3_server/checkpoints/mobilenet_server/model_mobile.json', 'r') as f:
    data = json.load(f)

# Trích xuất thông tin từ dữ liệu
epochs = [i for i in range(len(data))]
losses = [d["Loss"] for d in data]
accuracies = [d["Accuracy"] for d in data]

# Vẽ đồ thị
fig, ax1 = plt.subplots()

# Trục y cho Loss
color = 'tab:red'
ax1.set_xlabel('Epoch')
ax1.set_ylabel('Loss', color=color)
ax1.plot(epochs, losses, marker='o', color=color, label='Loss')
ax1.tick_params(axis='y', labelcolor=color)

# Trục y thứ 2 cho Accuracy
ax2 = ax1.twinx()
color = 'tab:blue'
ax2.set_ylabel('Accuracy', color=color)
ax2.plot(epochs, accuracies, marker='o', color=color, label='Accuracy')
ax2.tick_params(axis='y', labelcolor=color)

# Hoàn thiện đồ thị
plt.title('Loss and Accuracy per Epoch')
plt.grid(True)
fig.tight_layout()
plt.show()
```
